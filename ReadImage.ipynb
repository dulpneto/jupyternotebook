```python
%matplotlib inline
import matplotlib.pyplot as plt

"""setting the default colormap to gray"""
plt.rcParams['image.cmap'] = 'gray'
plt.rcParams['image.interpolation'] = 'none'
```

```python
from skimage import data

image_gray = data.camera()
plt.imshow(image_gray);
```

```python
import os
from skimage import io

filename = os.path.join('images', 'park.jpeg')

image = io.imread(filename)
plt.imshow(image);
```

```python
from skimage import color

image_gray = color.rgb2gray(image)
plt.imshow(image_gray);
```

```python
import warnings
from skimage import filters
from skimage.morphology import disk

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    smooth_image = filters.median(image_gray, disk(5))
plt.imshow(smooth_image);
```

```python
smooth_image = filters.gaussian(image_gray, 5)
plt.imshow(smooth_image);
```
